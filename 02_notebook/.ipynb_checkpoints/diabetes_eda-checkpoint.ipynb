<h2><b>Who are the Pima Indians?</b></h2>

The **Pima Indians**, historically known as **the Akimel O'odham ("River People")**, are a Native American tribe indigenous to the Gila and Salt River valleys in what is now central and southern Arizona, United States, and parts of Sonora, Mexico. They are widely recognized as **descendants of the prehistoric Hohokam culture**.

<h2><b>About the dataset</b></h2>

The dataset was gotten from [kaggle]( https://www.kaggle.com/datasets/jamaltariqcheema/pima-indians-diabetes-dataset) and it contains **769 entries** of which all are **female**, and **9 (nine) columns** which will be explain in details.

The objective of this project is to explore the dataset using NumPy, Pandas, Matplotlib, and Seaborn. We will analyze the correlation between key features (Pregnancy, Glucose level, BMI, Blood Pressure, Age, and Pedigree Function) and determine how they relate to the final outcome (i.e., whether a patient is diabetic or not).

<h2><b>Details</b></h2>

* **Pregnancies:** Number of times pregnant
* **Glucose:** Plasma Glucose Concentration a 2-hour in an Oral Glucose Tolerance Test (mg/dl).
* **BloodPressure:** Diastolic Blood Pressure (mm/Hg).
* **SkinThickness:** Triceps Skin Fold Thickness (mm)
* **Insulin:** 2-Hour Serum Insulin (mu U/ml)
* **BMI:** Body Mass Index (weight in kg/(height in m)^2)
* **DiabetesPedigreeFunction:** It provides information about diabetes history in relatives and genetic relationship of those relatives with patients. Higher Pedigree Function means patient is more likely to have diabetes.
* **Age:** Age of an individual (years)
* **Outcome:** Target Variable (0 or 1) where ‘0’ denotes patient is not diabetic and ‘1’ denotes patient is diabetic.

<h2><b>Attributes Normal Value Range</b></h2>

* **Glucose:** Glucose (< 140) = Normal, Glucose (140-200) = Pre-Diabetic, Glucose (> 200) = Diabetic
* **BloodPressure:** B.P (< 60) = Below Normal, B.P (60-80) = Normal, B.P (80-90) = Stage 1 Hypertension, B.P (90-120) = Stage 2 Hypertension, B.P (> 120) = Hypertensive Crisis
* **SkinThickness:** SkinThickness (< 10) = Below Normal, SkinThickness (10-30) = Normal, SkinThickness (> 30) = Above Normal
* **Insulin:** Insulin (< 200) = Normal, Insulin (> 200) = Above Normal
* **BMI:** BMI (< 18.5) = Underweight, BMI (18.5-25) = Normal, BMI (25-30) = Overweight, BMI (> 30) = Obese

<h2><b>What is Diabetes?</b></h2>

Diabetes is a chronic condition where the body either cannot produce enough insulin or cannot effectively use the insulin it makes. This causes sugar to build up in the bloodstream instead of being converted into energy. Over time, this high blood sugar can lead to serious complications like heart and kidney disease.

<h3><b>Types of Diabetes</b></h3>

* **Type 1:** An autoimmune reaction where the body mistakenly stops making insulin. It is usually diagnosed in children and young adults, and requires daily insulin intake.
* **Type 2:** The most common type (90-95% of cases), where the body's cells develop insulin resistance and cannot use insulin properly. It is heavily influenced by lifestyle factors and genetics.
* **Gestational Diabetes:** High blood sugar that develops during pregnancy in individuals who did not previously have diabetes. It usually goes away after the baby is born

**Keynote:** We will be analyzing data closely link to **Type 2 diabetes** where factors like high Glucose levels, BMI and Age serve as key predictive markers.

## Checking for missing values

Columns like Glucose, BloodPressure, SkinThickness, Insulin and BMI contains 0 values which serve as placeholder for missing values.

```python
# Identify entries with zero values
(df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] == 0).sum()
```

## Distribution of Glucose Levels across patients

* **Distribution Shape**: The glucose histogram is **positively skewed** (right-skewed), featuring a long tail extending toward the higher values.
* **Data Interpretation**: This indicates that the majority of patients cluster around normal or slightly elevated glucose ranges, while a minority of high-risk patients pull the distribution out with exceptionally elevated blood sugar levels.

## Distribution of Glucose Levels by Diabetic Outcome

```python
# set resolution and size
plt.figure(figsize=(8,5),dpi=100)

# Create Plot
ax = sns.histplot(data=df,x='Glucose',bins=40,hue='Outcome', multiple='stack')

# Rename title
sns.move_legend(ax, loc='best',title='Patient Status', labels=['Healthy','Diabetic'])

# Set title and labels
plt.title('Distribution of Glucose Levels by Diabetic Outcome',fontsize=14,pad=15)
plt.xlabel('Glucose Levels (mg/dl)',fontsize=14)
plt.ylabel('Count of patient',fontsize=14)

# Save the Plot
plt.savefig("C:/Users/X270/Downloads/Image/histplot2.png",dpi=300,bbox_inches='tight')

plt.show()
```

### Visual Insights: Stacked Glucose Distribution by Outcome
* **Low Glucose Range (<100 mg/dL)**: Healthy patients dominate the lower values. The proportion of diabetic patients in this range is extremely low, visually confirming that lower glucose levels correlate strongly with a non-diabetic status.
* **Overlapping Transition Zone (100–140 mg/dL)**: There is a significant mix of both classes in this middle band. This represents the borderline/prediabetic region where glucose alone cannot definitively separate the two groups.
* **High Glucose Range (>140 mg/dL)**: Diabetic patients overwhelmingly make up the bulk of the stacked bars as values increase. Beyond ~170 mg/dL, the bars become almost entirely diabetic.
* **Analytical Takeaway**: This plot visually justifies the positive skewness noted earlier. The long right-hand tail is heavily populated by the diabetic class, proving that an increasing glucose level is an exceptionally strong risk indicator.


## Interaction of BMI and Glucose levels by outcome

```python
df_sorted = df.sort_values('Outcome')
```

```python
# Set resolution and size
plt.figure(figsize=(8,5),dpi=150)

# Create the plot
ax = sns.scatterplot(data=df_sorted,x='BMI',y='Glucose',hue='Outcome',edgecolor=None,alpha=0.7)

# Customize Legend
sns.move_legend(ax,loc='upper left',bbox_to_anchor =(1.02,1),title='Patient Status', labels=['Healthy','Diabetic'])

# Set title and label
plt.title('Interaction of BMI and Glucose Levels by diagnosis',fontsize=14,pad=15)
plt.xlabel('BMI ($Kg/m^2$)',fontsize=12)
plt.ylabel('Glucose Level(mg/dl)',fontsize=12)

# Save the Plot
plt.savefig("C:/Users/X270/Downloads/Image/scatterplot.png",dpi=300,bbox_inches='tight')

# Show Plot
plt.show()
```

### Visual Insights: BMI vs. Glucose Scatter Plot Matrix
* **The High-Risk Cluster (Top-Right)**: There is a heavy concentration of diabetic patients in the upper-right quadrant of the plot where both BMI exceeds 30 kg/m² and Glucose exceeds 140 mg/dL. 
* **The Healthy Cluster (Bottom-Left)**: Lower-left regions feature a dominant mix of healthy patients, demonstrating that lower weights combined with normal blood sugar strongly indicate non-diabetic profiles.
* **Feature Interaction**: Neither feature operates perfectly in isolation; some patients have an obese BMI (>30 kg/m²) but remain healthy due to normal glucose, while others have high glucose but lower BMIs. 
* **Analytical Takeaway**: Neither Glucose nor BMI tells the full story alone. However, combining them creates a distinct diagonal boundary that cleanly separates healthy patients from diabetic patients, making them powerful joint predictors.

### Visual Insights: Glucose Distribution Boxplot

* **Interquartile Range (IQR)**: The middle 50% of the diabetic group sits completely higher than the middle 50% of the healthy group, showing minimal overlap between their core distributions.
* **Outliers**: The healthy group contains several high-value outliers (the dots above the top whisker). This reveals that while high glucose is typical for diabetes, some healthy individuals can also experience elevated spikes.

```python
# Set size and resolution
plt.figure(figsize=(9, 5), dpi=100)

# Create a bar plot showing the average outcome (diabetes rate) for each pregnancy count
sns.barplot(data=df, x='Pregnancies', y='Outcome', errorbar=None,color='#F71919')

# Add a reference line for the overall average dataset risk (~35%)
plt.axhline(y=df['Outcome'].mean(),color='black', linestyle='--',alpha=0.7,label='Average Dataset Risk')

# Add Title and label
plt.title('Diabetes Risk Rate by Number of Pregnancies', fontsize=14, pad=15)
plt.xlabel('Number of Pregnancies', fontsize=12)
plt.ylabel('Probability of Diabetes (Outcome Rate)', fontsize=12)
plt.legend()
# save the Plot
plt.savefig("C:/Users/X270/Downloads/Image/barplot2.png",dpi=300,bbox_inches='tight')

plt.show()
```

* **The Baseline Comparison:** The horizontal dashed line (plt.axhline) serves as a benchmark. It instantly shows  where the risk switches from "below average" to "above average".
* **Clear Risk Escalation:** For low pregnancy counts (0, 1, and 2), the red bars sit safely below the dashed line. Starting around 3 or 4 pregnancies, the risk crosses the line and stays significantly higher than the dataset average. 
* **Extreme Risk in High Counts:** For patients with 7 to 14+ pregnancies, the probability bars climb toward 0.6 to 0.8, meaning an individual in that group has a 60% to 80% chance of being diabetic.

**Keynote**
* **Pregnancy is a Compounding Risk Factor:** The plot proves that the probability of having diabetes increases directly with the number of pregnancies.

### 📘 Final Conclusion & Key Insights

Through this Exploratory Data Analysis (EDA) using Pandas and Seaborn, we successfully cleaned, analyzed, and visualized the patient health dataset. The key takeaways from our investigation include:

#### Data Integrity & Cleaning
*  Physically impossible `0` values in continuous medical features (`Glucose`, `BMI`, `BloodPressure`) were correctly identified as missing records and masked as `NaN` to preserve statistical accuracy.
*  Heavily corrupted variables (`Insulin` and `SkinThickness`) were entirely removed to prevent introducing heavy statistical bias into future analyses.

####  Target Distribution
* The dataset exhibits a 65/35 mild-to-moderate class imbalance skewed toward healthy patients.
* While safe for visualization and exploration, any future machine learning deployment must use **Precision, Recall, and F1-Score** rather than baseline accuracy to evaluate performance.

####  Primary Health Risk Drivers
* **Glucose Level:** This emerged as the single most discriminative feature in the dataset. Diabetic boxes and distributions sat significantly higher than healthy ones, with a strong conversion rate (~69%) to diabetes once glucose surpassed 140 mg/dL.
* **BMI:** This acts as a powerful secondary physical indicator, though it shows a higher degree of vertical distribution overlap between target classes compared to Glucose.

####  Underlying & Compounding Risk Factors
* **Age & Life Stages:** Age exhibits a weak-to-moderate linear correlation (~0.17) due to its non-linear behavior and a younger sample population bias. However, it acts as a critical threshold factor; patients under 30 are predominantly healthy, whereas a distinct risk inversion occurs past age 50, where diabetic outcomes completely dominate the older cohort.
* **Pregnancy History:** The number of pregnancies acts as a compounding metabolic stressor. While a history of 0 to 2 pregnancies keeps a patient's risk profile safely below the dataset baseline (~35%), crossing a threshold of 3 to 4 pregnancies shifts risks above average, escalating to a 60%–80% probability of diabetes in high-order pregnancy counts (7+).
* **Genetic Predisposition:** The `DiabetesPedigreeFunction` boxplot establishes genetics as a significant baseline risk multiplier. Diabetic patients display a noticeably higher median pedigree score and a denser clustering of high-value outliers, confirming that a rich family history lowers the body's threshold for resisting the disease.
* **Blood Pressure Limitations:** In contrast to the other metrics, diastolic `BloodPressure` (once cleared of zero-value errors) displayed highly overlapping distributions between healthy and diabetic cohorts and a very weak direct correlation (~0.06 to 0.12). It serves as a poor independent predictor, functioning instead as a co-occurring metabolic symptom rather than a primary diagnostic driver.

####  Feature Interactions
*  Bivariate analysis demonstrated that Glucose and BMI work together as an effective pair, providing a clearer boundary for classification when viewed together on a scatter plot than when analyzed separately.
*  The correlation matrix confirmed that the remaining columns—including underlying drivers like Age, Pregnancies, and Genetics—provide completely unique, independent, and non-redundant clues, making them ideal joint inputs for a future predictive model.
